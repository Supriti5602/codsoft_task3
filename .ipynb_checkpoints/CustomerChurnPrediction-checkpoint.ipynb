## Encode the categorical data

```python
dataset.select_dtypes(include='object').columns
```

```python
dataset = dataset.drop(columns=['RowNumber','CustomerId','Surname'])
```

```python
# Print column names
print(dataset.columns)

# Print column data types
print(dataset.dtypes)
```

```python
numerical_columns = ['CreditScore', 'Age', 'Tenure', 'Balance',
       'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary',
       'Exited']
dataset.groupby('Geography')[numerical_columns].mean()
```

```python
dataset.groupby('Gender')[numerical_columns].mean()
```

```python


# Assuming you have a DataFrame named 'dataset'

# Perform one-hot encoding with drop_first=True
dataset = pd.get_dummies(data=dataset, drop_first=True)

# Find boolean columns
boolean_columns = dataset.select_dtypes(include=['bool']).columns

# Convert boolean columns to integers (True -> 1, False -> 0)
dataset[boolean_columns] = dataset[boolean_columns].astype(int)

# Print the modified dataset
dataset.head()
```

## CountPlot

```python
# Assuming 'Exited' is the column you want to plot
sns.countplot(data=dataset, x='Exited')
plt.show()
```
